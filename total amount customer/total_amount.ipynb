```python
import findspark
findspark.init()
from pyspark import SparkConf, SparkContext

conf = SparkConf().setMaster("local").setAppName("FriendsByAge")
sc = SparkContext(conf = conf)
```

```python
lines = sc.textFile("file:///SparkCourse/total amount customer/customer-orders.csv")
def key_value_method(x):
    splited = x.split(",")
    id = int(splited[0])
    spent = float(splited[2])
    return (id, spent)

key_value = lines.map(key_value_method)
total_rdd = key_value.reduceByKey(lambda x,y :x+y).sortBy(lambda x:x[1])
total_list = total_rdd.collect()

for key, value in total_list:
    print("Customer #{} spend {} in total.".format(key, round(value,2)))
```
